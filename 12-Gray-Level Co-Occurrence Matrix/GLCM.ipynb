```python
import matplotlib.pyplot as plt
import numpy as np
import skimage
from skimage.feature import graycomatrix, graycoprops
```

```python
B5 = skimage.io.imread("E:/Landsat8/NIR.tif")
```

```python
np.shape(B5)
```

```python
result = np.empty((998,998), dtype= float)

image = np.array(B5, dtype='uint8')
```

```python
window_size = 3
```

```python
for i in range(0,998):
    for j in range(0,998):
        window = image[i : i + window_size, j:j+ window_size]
        glcm = graycomatrix(window, [1], [0], normed=True, symmetric=True)
        res = graycoprops(glcm , 'contrast')[0][0]
        result[i,j] = res
plt.figure(figsize=(15,10))
plt.imshow(result)
plt.show()
```
